### Store Part I results into DataFrame

```python
city_df = pd.read_csv("city_data.csv")
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Access maps with unique API key
gmaps.configure(api_key=g_key)
```

```python
coords = city_df[["Latitude","Longitude"]]
humidity = city_df["Humidity"]
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(coords, weights=humidity, 
                                 dissipating=False, max_intensity=10,
                                 point_radius=1)
```

```python
# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
city_df = city_df.dropna()
hotels_df = city_df[ (city_df["Humidity"]<65) &
                        (city_df["Temperature"]>65) &
                        (city_df["Temperature"]<75) &
                        (city_df["Cloudiness"]<70) &
                        (city_df["Wind Speed"]<10)]
len(hotels_df)
hotels_df["Hotel Name"] = ""
hotels_df.head()
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# geocoordinates
target_search = "hotel"
target_radius = 5000
target_type = "hotel"

# set up a parameters dictionary
params = {
    "keyword": target_search,
    "radius": target_radius,
    "type": target_type,
    "key": g_key
}
```

```python
# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
```

```python
for index, row in hotels_df.iterrows():
    params["location"] = f'{row["Latitude"]},{row["Longitude"]}'
    try: 
        response = requests.get(base_url, params=params)
        data = response.json()
        hotels_df.loc[index,8] = data["results"][0]["name"]
    except:
        print(f"No hotel found for search query: {row['City']}")
        pass
```
